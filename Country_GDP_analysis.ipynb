```python
import matplotlib.pyplot as plt # visualisation
import seaborn as sns #distribution visualization(distribution plot, line plot)

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2

import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
df['InternetUsers']
```

```python
#Distributions:
vis1 = sns.distplot(df['InternetUsers']) #UNIVARIABLE ANALYSIS= plot the graph using one variable
plt.show()
```

```python
vis2 = sns.displot(df['InternetUsers'])
plt.show()
```

```python
vis3 = sns.distplot(df['InternetUsers'], bins=10)
plt.show()
```

```python
#BOX PLOT
vis4 = sns.boxplot(data = df, x= 'IncomeGroup', y= "BirthRate") #BI VARIABLE ANALYSIS
plt.show()
```

Outlier= in statistic outlier is the datapoints which is very far from other datapoints
we detect outlier by graphs ie boxplot


```python
vis5 = sns.lmplot(data = df, x= 'InternetUsers', y= "BirthRate")
plt.show()
```

```python
vis6 = sns.lmplot(data = df, x= 'InternetUsers', y= "BirthRate", fit_reg= False)
plt.show()
```

```python
vis6 = sns.lmplot(data = df, x= 'InternetUsers', y= "BirthRate", fit_reg= True)
plt.show()
```

```python
vis6 = sns.lmplot(data = df, x= 'InternetUsers', y= "BirthRate", fit_reg= False, hue= "IncomeGroup") #hue parameter for color
plt.show()
```

```python
vis7 = sns.lmplot(data = df, x= 'InternetUsers', y= "BirthRate", fit_reg= True, hue= "IncomeGroup") #hue parameter for color
plt.show()
```
